```python
'''
Imagine you are working for an e-commerce company and the marketing team wants to target customers more
effectively by creating segments based on their purchasing behavior. However, traditional segmentation 
methods are not providing satisfactory results.
'''
#Solution: (Density based spatial clustering of application with noise)
# Use DBSCAN to identify clusters of customers with similar purchasing patterns.
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
data = np.random.rand(100,2)
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)
```

```python
#apply DBSCAN
dbscan = DBSCAN(eps=0.3,min_samples=5)
cluster = dbscan.fit_predict(scaled_data)
```

```python
# visualaize
plt.scatter(data[:,0],data[:,1],c=cluster,cmap='viridis',marker='o',s=50)
plt.title("DBSCAN Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
result_df = pd.DataFrame({'Feature 1':data[:,0],'Feature 2':data[:,1],'Cluster':cluster })
print("Size of Cluster: ",len(np.unique(cluster)))
print("Size of Each Cluster")
print(result_df['Cluster'].value_counts())
```

```python
# model testing input
user_input = np.array([[0.6,0.8]])
user_input_scaler = scaler.transform(user_input)
result = dbscan.fit_predict(user_input_scaler)
print("User blong to cluster: ",result[0])
result
```

```python
'''
This segmentation will help you tailor marketing strategies 
and promotions to different customer groups, ultimately maximizing
sales and customer satisfaction
'''
# Gausion Mixture model (GMM)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import StandardScaler
```

```python
np.random.seed(42)
segmentation1 = np.random.normal(loc=30, scale=5,size=100)
segmentation2 = np.random.normal(loc=60, scale=10,size=150)
segmentation3 = np.random.normal(loc=90, scale=8,size=120)
data = np.concatenate([segmentation1,segmentation2,segmentation3]).reshape(-1,1)
```

```python
scaler = StandardScaler()
scaler_data = scaler.fit_transform(data)
```

```python
n_components=3
GMM = GaussianMixture(n_components=n_components,random_state=42)
fit = GMM.fit(scaler_data)
fit
```

```python
#prediction
cluster_label = GMM.predict(scaler_data)
#visualize
plt.scatter(data,np.zeros_like(data),c=cluster_label,cmap='viridis')
plt.title("The Gaussian Mixture of Model")
plt.xlabel("Purchasing Amount")
plt.show()
```

```python
user_input = float(input("ENTER purchasing amount: "))
user_input_scaler= scaler.transform(np.array([[user_input]]))
predict_input= GMM.predict(user_input_scaler.reshape(-1,1))[0]
print(f"The user input {user_input} and purchasing prediction: {predict_input+1} ")
```
